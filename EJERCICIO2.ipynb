**CALCULADORA** + - * /

```python
num1 = int(input("Digite el número 1: "))
num2 = int(input("Digite el número 2: "))

suma = num1 + num2
resta = num1 - num2
multi = num1 * num2
divi = num1 / num2

print("\n EL resultao de la SUMA es= {}".format(suma))
print("\n EL resultao de la RESTA es= {}".format(resta))
print(f"\n EL resutlado de la MULTIPLICACIÓN es= {multi}")
print(f"\n EL resutlado de la DIVISIÓN es= {divi}")
```

Se recibe una venta por concepto de un prodcuto se requiere aplicar el IVA (19%) y conocer el total a pagar con el IVA Incluido

```python
# variable para capturar la venta
venta = float(input("¿Cuanto fue el total de la compra?: "))
# Cálculo del iva
iva = venta * 0.19 
# suma del iva con la venta
total = venta + iva
# Muestro por pantalla el total con el iva
print("EL IVA es= ${:,.0f} y el total a pagar es=${:,.0f}".format(iva, total ))
```

Se requiere conocer el promerdio general de una materia que está dada por tres calificaciones


```python
cal1 = float(input("Digite la calificaicón 1: "))
cal2 = float(input("Digite la calificaicón 2: "))
cal3 = float(input("Digite la calificaicón 3: "))

prom = round((cal1 + cal2 + cal3) / 3, 2)

print(f"El promedio es= {prom}")
```
